```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
days = np.arange(1, 301)
```

```python
prices = 0.05 * days + 100 + np.cumsum(np.random.normal(0,1,300))
```

```python
data = pd.DataFrame({'Day': days, 'Price':prices})
```

```python
plt.figure(figsize=(10, 4))
plt.plot(data['Day'], data['Price'], label='Stock Price')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Synthetic Stock Prices with Trend')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X = data[['Day']]
y = data['Price']
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2,random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):",round(mse,2))
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(X_test[:, 1], y_test, color='blue', label='Actual')
plt.scatter(X_test[:, 1], y_pred, color='red', label='Predicted')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Actual vs Predicted Prices (Polynomial Regression)')
plt.legend()m+


plt.grid(True)
plt.show()
```
